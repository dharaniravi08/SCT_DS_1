```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
population_data=pd.read_csv("C:/Users/P3Fusion/Downloads/archive (3)/World Marriage Dataset.csv")
print(population_data.head())
plt.figure(figsize=(12, 6))
sns.countplot(data=population_data, x='gender', color='orange')
plt.title('gender Distribution in the World Population')
plt.xlabel('gender')
plt.ylabel('population')
plt.grid(True)
plt.tight_layout() 
plt.show()
```
